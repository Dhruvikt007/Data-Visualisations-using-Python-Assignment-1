### Question 1
Plot using following plots on fmri data set:
* sns.relplot 
* line plot
* box plot



```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

# Using sns.relplot

```python
sns.relplot(
    
    x = 'timepoint', y = 'signal', 
    data = fmri, hue = 'subject', 
    size = 'event', style = 'region', 
    col = 'region', row = 'event'
)
```

# Using Lineplot

```python
D = sns.relplot(
    data=fmri,
    x='timepoint', y='signal',
    hue='event', style='event', col='region',
    height=5, aspect=1, kind='line'
)

(D.set_axis_labels("Timepoint", "Percent Signal Change")
  .set_titles("Region: {col_name} Cortex")
  .tight_layout(w_pad=0))
```

# Using Boxplot

```python
sns.catplot(x="timepoint", y="signal",
                hue="region", col="event",
                data=fmri, kind="box",
                height=5, aspect=1);
```
